```python
import joblib
import numpy as np

encoder = joblib.load('encoder.pkl')
model = joblib.load('model.pkl')
team_rank = joblib.load('team_rank_dict.pkl')
```

```python
def predict_match_winner(home_team,away_team,is_netu=True):
    home_rank = team_rank.get(home_team ,50)
    away_rank = team_rank.get(away_team,50)

    rank_diff = home_rank - away_rank
    net_val = 1 if is_netu else 0 

    match_features = np.array([[rank_diff,net_val]],dtype=float)

    pred_encoded = model.predict(match_features)[0]
    team = encoder.inverse_transform([pred_encoded])[0]

    if team == "home":
        print(f'the winner will be {home_team}')
    elif team == 'away':
        print(f"the winner will be {away_team}")
    else:
        print("draw")

    return team
```

```python
predict_match_winner(home_team='Egypt',away_team='Belgium')
```

```python
joblib.dump(predict_match_winner,'predict.pkl')
```
